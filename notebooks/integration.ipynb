```python
import pandas as pd
import os
```

```python
os.makedirs("data/processed", exist_ok=True)

cdc = pd.read_csv("data/clean/cdc_cleaned.csv")
atlas = pd.read_csv("data/clean/foodatlas_clean.csv")
```

```python
merged = pd.merge(cdc, atlas, on="FIPS", how="inner")

merged.to_csv("data/processed/merged_output.csv", index=False)

print("Merged dataset saved to data/processed/merged_output.csv")
```
