```python
import pandas as pd
import os
```

```python

os.makedirs("data/processed", exist_ok=True)

cdc = pd.read_csv("data/clean_1/cdc_cleaned.csv")
atlas = pd.read_csv("data/clean_1/foodatlas_clean.csv")
```

```python
merged = pd.merge(cdc, atlas, on="FIPS", how="inner")

merged.to_csv("data/merge/merged_output.csv", index=False)

print("Merged dataset saved to data/merge/merged_output.csv")
```
