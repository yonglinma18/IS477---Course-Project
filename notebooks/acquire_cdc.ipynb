```python
import os
import requests
import hashlib
```

# CDC dataset URL and output name


```python
os.makedirs("data/raw", exist_ok=True)

cdc_url = "https://uofi.box.com/shared/static/kyr7iscrhy6943ictwt7voadier5mf7s"
cdc_filename = "data/raw/cdc_places_2024.csv"
```

# Download CDC dataset


```python
response = requests.get(cdc_url)
with open(cdc_filename, "wb") as f:
    f.write(response.content)

print("Saved CDC dataset:", cdc_filename)
```

# Compute SHA-256


```python
def sha256(path):
    with open(path, "rb") as f:
        return hashlib.sha256(f.read()).hexdigest()

print("CDC SHA-256:", sha256(cdc_filename))
```
